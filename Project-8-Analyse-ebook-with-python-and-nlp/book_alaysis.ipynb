# Load the Book

```python
with open("283 - Get Number of Chapters - miracle_in_the_andes.txt")as file:
    book = file.read()
```

```python
type(book)
```

# How many Chapters

### With String Method

```python
book.count("Chapter")
```

### with regular expressions

```python
import re
pattern = re.compile("Chapter [0-9]+")
findings = re.findall(pattern,book)
len(findings)
```

### which are the sentence where "Love" is present

```python
# pattern = re.compile("""[a-zA-Z ,!:;'"]* love [a-zA-Z ,!:;'"]*""")
pattern = re.compile('[^.]*love[^a-zA-Z]+[^.]*.')
finding = re.findall(pattern,book)
len(finding)
```
